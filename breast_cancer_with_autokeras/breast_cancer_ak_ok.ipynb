```python
import autokeras as ak
import tensorflow as tf
```

## Importando DataSet

```python
data_test = pd.read_csv('https://query.data.world/s/rit43ay2hzyhvtpfvv43vaklcqdzfo')
data_train= pd.read_csv('https://query.data.world/s/tyqrpeyl4ensbl2xtghsfjkandkn46')
```

```python
data_test.head()
```

## Separando el target del dataset

```python
y_test = data_test['Target']
y_train = data_train['Target']

x_test = data_test.drop('Target', axis = 1)
x_train = data_train.drop('Target', axis = 1)
```

```python
x_train.head()
```

```python
x_train.info()
```

```python
x_test.info()
```

## No hay elementos nulos

```python
x_train = x_train.to_numpy(dtype='float', na_value=np.nan)
x_test = x_test.to_numpy(dtype='float', na_value=np.nan)
y_train = y_train.to_numpy(dtype='float', na_value=np.nan)
y_test = y_test.to_numpy(dtype='float', na_value=np.nan)
```

```python
x_train
```

## Creando y entrenando el modelo con autokeras

```python
input_node =  ak.StructuredDataInput()

output_node_1 = ak.Normalization()(input_node)

output_node_2 = ak.DenseBlock()(output_node_1)

output_node_3 = ak.ClassificationHead(loss = 'binary_crossentropy', metrics=['accuracy'])(output_node_2)


cbs = [tf.keras.callbacks.EarlyStopping(patience=2)]


model = ak.AutoModel(input_node , output_node_3, overwrite=True, max_trials=10)


model_ = model.fit(x_train , y_train, callbacks = cbs , validation_split = 0.2 , batch_size = 16, epochs=20)
```

## Evaluaando el modelo en el test set

```python
model.evaluate(x_test, y_test)
```

## El modelo tiene una precision de 99.4 %

## Guardando el modelo

```python
model.export_model().save('breast_cancer_ak_model.h5')
```

## Cargando el modelo

#### Es importante tener en cuenta que debes cargar los objetos personalizados ak.CUSTOM_OBJECTS al cargar el modelo, ya que el modelo originalmente entrenado con AutoKeras utiliza estos objetos personalizados para definir su arquitectura.

```python
from keras.models import load_model

# Cargar el modelo guardado
loaded_model = load_model('breast_cancer_ak_model.h5', custom_objects=ak.CUSTOM_OBJECTS)
```

```python
loaded_model.weights
```

### Guardando los pesos del modelo

```python
loaded_model.save_weights('breast_cancer_weights.h5')
```
